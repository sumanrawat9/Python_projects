```python
import cv2
import numpy as np
import pandas as pd
```

```python
colors= pd.read_csv('colors.csv')
colors
```

```python
def get_color_name(R, G, B):
    minimum = 10000
    cname = ""
    for i in range(len(colors)):
        d = abs(R - int(colors.loc[i, "R"])) + abs(G - int(colors.loc[i, "G"])) + abs(B - int(colors.loc[i, "B"]))
        if d <= minimum:
            minimum = d
            cname = colors.loc[i, "color_name"]
    return cname
```

```python
def classify_color(image_path):
    # Read the image
    img = cv2.imread(image_path)
    img = cv2.resize(img, (300, 300))

    # Convert image to RGB (OpenCV uses BGR by default)
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    # Calculate the average color of the image
    avg_color_per_row = np.average(img_rgb, axis=0)
    avg_color = np.average(avg_color_per_row, axis=0)
    avg_color = avg_color.astype(int)

    # Get the color name
    color_name = get_color_name(avg_color[0], avg_color[1], avg_color[2])
    return color_name, avg_color
```

```python
image_path = '/content/img.png'
color_name, avg_color = classify_color(image_path)
print(f"The color is {color_name} with RGB values {avg_color}")
```
